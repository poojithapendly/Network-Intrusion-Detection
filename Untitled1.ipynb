```python
import pandas as pd

columns = [
    'duration', 'protocol_type', 'service', 'flag', 'src_bytes', 'dst_bytes',
    'land', 'wrong_fragment', 'urgent', 'hot', 'num_failed_logins', 'logged_in',
    'num_compromised', 'root_shell', 'su_attempted', 'num_root', 'num_file_creations',
    'num_shells', 'num_access_files', 'num_outbound_cmds', 'is_host_login',
    'is_guest_login', 'count', 'srv_count', 'serror_rate', 'srv_serror_rate',
    'rerror_rate', 'srv_rerror_rate', 'same_srv_rate', 'diff_srv_rate',
    'srv_diff_host_rate', 'dst_host_count', 'dst_host_srv_count', 'dst_host_same_srv_rate',
    'dst_host_diff_srv_rate', 'dst_host_same_src_port_rate', 'dst_host_srv_diff_host_rate',
    'dst_host_serror_rate', 'dst_host_srv_serror_rate', 'dst_host_rerror_rate',
    'dst_host_srv_rerror_rate', 'attack_type', 'difficulty_level'
]

df = pd.read_csv("kdd_train.csv", names=columns,skiprows=1,low_memory=False)

df.head()
```

```python
from sklearn.preprocessing import StandardScaler

# --- 1. BINARY LABELING (Normal vs. Attack) ---
# If the attack_type is 'normal', give it a 0 (Safe). Otherwise, give it a 1 (Danger!).
df['label'] = df['attack_type'].apply(lambda x: 0 if x == 'normal' else 1)

# Drop the old text attack columns so they don't confuse the model
df = df.drop(['attack_type', 'difficulty_level'], axis=1)


# --- 2. ONE-HOT ENCODING (Turning text into columns of 0s and 1s) ---
# Identify columns that contain text words
text_columns = ['protocol_type', 'service', 'flag']

# pd.get_dummies automatically splits text columns into multiple true/false numeric columns
df_encoded = pd.get_dummies(df, columns=text_columns, drop_first=True)


# --- 3. SPLITTING DATA & FEATURE SCALING ---
# Separate our clues (X) from the final answer label (y)
X = df_encoded.drop(['label'], axis=1)
y = df_encoded['label']

# Convert any remaining True/False values to 1 and 0 so everything is strictly a number
X = X.astype(float)

# Scale the features so big numbers don't overpower small numbers
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Data preprocessing complete!")
print("Cleaned data shape:", X_scaled.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

# --- 1. SPLIT DATA INTO WORKBOOK (TRAIN) & EXAM (TEST) ---
# We save 20% of our data for the final exam (test_size=0.2)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

print(f"Training set size: {X_train.shape[0]} samples")
print(f"Testing set size: {X_test.shape[0]} samples")


# --- 2. INITIALIZE THE MACHINE LEARNING BRAIN ---
# We pick a Decision Tree Classifier
ai_guard = DecisionTreeClassifier(max_depth=5, random_state=42)


# --- 3. TRAIN THE MODEL (The Learning Process) ---
# .fit() means "study these clues (X_train) and their correct answers (y_train)"
ai_guard.fit(X_train, y_train)

print("Training complete! The AI guard has finished studying.")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report

# --- 1. TAKE THE EXAM ---
# We give the AI the test clues (X_test) and ask it to guess the answers
y_pred = ai_guard.predict(X_test)


# --- 2. CALCULATE THE SCORES ---
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)


# --- 3. PRINT THE RESULTS ---
print("=== AI GUARD REPORT CARD ===")
print(f"Accuracy : {accuracy * 100:.2f}%")
print(f"Precision: {precision * 100:.2f}%")
print(f"Recall   : {recall * 100:.2f}%")
print("\n--- Confusion Matrix ---")
print(f"True Normals (Correctly Let In) : {cm[0][0]}")
print(f"False Alarms (Good guys blocked): {cm[0][1]}")
print(f"Missed Thieves (Bad guys let in): {cm[1][0]}")
print(f"True Attacks (Correctly Blocked): {cm[1][1]}")
```

```python
import numpy as np

def live_traffic_scanner(raw_connection_data):
    """
    Simulates our AI checking a single, live network connection.
    Takes a raw dictionary of connection data, matches it to our 
    trained workspace, and returns a live security alert.
    """
    
    # 1. Create a blank row with all columns initialized to 0.0
    # This acts as our empty template matching our training columns structure
    encoded_sample = pd.DataFrame(0.0, index=[0], columns=X.columns)
    
    # 2. Fill in the numeric features provided by the live system
    for feature in raw_connection_data:
        if feature in encoded_sample.columns:
            encoded_sample[feature] = float(raw_connection_data[feature])
            
    # 3. Handle the text/categorical variables (Manual One-Hot Encoding matching)
    # If the live connection is 'tcp', we flip the 'protocol_type_tcp' column from 0 to 1
    if 'protocol_type' in raw_connection_data:
        proto_col = f"protocol_type_{raw_connection_data['protocol_type']}"
        if proto_col in encoded_sample.columns:
            encoded_sample[proto_col] = 1.0
            
    if 'service' in raw_connection_data:
        serv_col = f"service_{raw_connection_data['service']}"
        if serv_col in encoded_sample.columns:
            encoded_sample[serv_col] = 1.0
            
    if 'flag' in raw_connection_data:
        flag_col = f"flag_{raw_connection_data['flag']}"
        if flag_col in encoded_sample.columns:
            encoded_sample[flag_col] = 1.0

    # 4. Scale the live sample using the exact same weights from our workshop
    scaled_sample = scaler.transform(encoded_sample)
    
    # 5. Ask our trained AI Guard to make a snap prediction
    prediction = ai_guard.predict(scaled_sample)[0]
    
    # 6. Output the easy-to-read decision text
    if prediction == 1:
        return "🚨 ALERT: Intrusion/Attack Detected! Connection Blocked."
    else:
        return "✅ SAFE: Normal Traffic. Connection Allowed."

# ========================================================
# 🧪 TEST RUN: Let's simulate a live connection arriving!
# ========================================================

# This sample connection looks normal: short duration, standard HTTP web traffic
incoming_visitor = {
    'duration': 0,
    'protocol_type': 'tcp',
    'service': 'http',
    'flag': 'SF',
    'src_bytes': 250,
    'dst_bytes': 1200,
    'count': 1,
    'srv_count': 1
}

# Run the live connection through our robot guard
security_status = live_traffic_scanner(incoming_visitor)
print(security_status)
```
